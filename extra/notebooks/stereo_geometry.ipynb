# Stereo geometry
### Disparity, distance, baseline of a stereocamera setup

See [Willow Garage - Stereo Geometry](http://pub1.willowgarage.com/~konolige/svs/disparity.htm)

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from IPython.html.widgets import interact, fixed, FloatSliderWidget, interactive
```

```python
def plot_disparity_by_range(baseline_mm, focal_length_mm, pixel_size_um, min_dist_mm=500,  max_dist_mm = 20000):
    x = np.linspace(min_dist_mm, max_dist_mm, 1000)
    #  bf / d 
    y = (baseline_mm * focal_length_mm) / (pixel_size_um / 1000 * x)
    plt.plot(x, y)
    plt.title('Stereo Disparity - b*f/(p*r)');
    plt.xlabel('r - range in mm')
    plt.ylabel('d - disparity in pixels')

def plot_range_resolution_log(baseline_mm, focal_length_mm, pixel_size_um, min_dist_mm=500, max_dist_mm = 10000):
    x = np.linspace(min_dist_mm, max_dist_mm, 1000)
    #  r ^2 / (bf)
    y =  x**2 * (pixel_size_um/1000) / (baseline_mm * focal_length_mm)
    logy = np.log10(y)
    plt.plot(x, logy)
    plt.title('Range Resolution -  delta r = (r2/bf) delta d');
    plt.xlabel('r - range in mm')
    plt.ylabel('range resolution in mm - log10 scale')
    pass
    

```

```python
# from example
plot_disparity_by_range(baseline_mm=80, focal_length_mm=6.3, pixel_size_um=14)
```

```python
# from example
plot_range_resolution_log(baseline_mm=80, focal_length_mm=6.3, pixel_size_um=14)
```

```python
# team diana setup
interactive(plot_disparity_by_range, baseline_mm=FloatSliderWidget(min=10, max=300, step = 1, value =100), \
                                                           focal_length_mm=fixed(6),  \
                                                           pixel_size_um=fixed(5.3), \
                                                           min_dist_mm=fixed(1000), \
                                                           max_dist_mm=fixed(2000))
```

```python
# team diana setup
interactive(plot_range_resolution_log, baseline_mm=FloatSliderWidget(min=10, max=300, step = 1, value =100), \
                                                             focal_length_mm=fixed(6),  \
                                                             pixel_size_um=fixed(5.3), \
                                                             min_dist_mm=fixed(1000), \
                                                             max_dist_mm=fixed(10000)) 
```
